```python
import cobra
```

```python
from cobra import *
```

```python
from escher import Builder
```

```python
model = Model('test')
```

```python
###### =====>GLU
GLUin = Reaction('GLUin')
GLUin.name='Glucose import'
GLUin.lower_bound=1
GLUin.upper_bound=1

##### Gluc+ATP =========> G6P+ADP+H
HK=Reaction('HK')
HK.name='Hixokinasis'
HK.lower_bound=0
HK.upper_bound=1000

##### G6P <=========> F6P
PGI=Reaction('PGI')
PGI.name='Glucose 6-phosphate isomerase'
PGI.lower_bound=-1000
PGI.upper_bound=1000

#####F6P+ATP=======>FDP+ADP+H
PFK=Reaction('PFK')
PFK.name='Phosphofructokinase'
PFK.lower_bound=0
PFK.upper_bound=1000

#####DHAP<===========>G3P
TPI=Reaction('TPI')
TPI.name='Triose-phosphate isomerase'
TPI.lower_bound=-1000
TPI.upper_bound=1000

#####FDP<===========>DHAP+G3P
ALD=Reaction('ALD')
ALD.name='Fructose 1,6-diphosphate aldolase'
ALD.lower_bound=-1000
ALD.upper_bound=1000

#####G3P+NAD+P<===========>DPG13+NADH+H
GAPDH=Reaction('GAPDH')
GAPDH.name='Glyceraldehyde 3-phosphate dehydrogenase'
GAPDH.lower_bound=-1000
GAPDH.upper_bound=1000

#####DPG13 + ADP<===========>PG3 + ATP
PGK=Reaction('PGK')
PGK.name='Phosphoglycerate kinase'
PGK.lower_bound=-1000
PGK.upper_bound=1000

#####PG3<===========>PG2
PGLM=Reaction('PGLM')
PGLM.name='Phosphoglycerate mutase'
PGLM.lower_bound=-1000
PGLM.upper_bound=1000

#####PG2<===========>PEP + H2O
ENO=Reaction('ENO')
ENO.name='Enolase'
ENO.lower_bound=-1000
ENO.upper_bound=1000

#####PEP + ADP + H ===========>PYR + ATP
PK=Reaction('PK')
PK.name='Pyruvate kinase'
PK.lower_bound=0
PK.upper_bound=1000

#####PYR + NADH + H ===========>LAC + NAD
LDH=Reaction('LDH')
LDH.name='Lactate dehydrogenase'
LDH.lower_bound=-1000
LDH.upper_bound=1000

#####AMP ========>
AMPe=Reaction('AMPe')
AMPe.name='AMP export'
AMPe.lower_bound=0
AMPe.upper_bound=1000
              
#####2ADP <========> AMP + ATP
APK=Reaction('APK')
APK.name='Adenylate kinase'
APK.lower_bound=-1000
APK.upper_bound=1000

#####PYR <========> 
PYRe=Reaction('PYRe')
PYRe.name='Pyruvate exchange'
PYRe.lower_bound=-1000
PYRe.upper_bound=1000

#####LAC <========> 
LACe=Reaction('LACe')
LACe.name='Lactate exchange'
LACe.lower_bound=-1000
LACe.upper_bound=1000

#####ATP + H2O ========> ADP + P + H
ATPh=Reaction('ATPh')
ATPh.name='ATP hydrolysis'
ATPh.lower_bound=0
ATPh.upper_bound=1000

#####NADH ========> NAD + H
NADHo=Reaction('NADHo')
NADHo.name='NADH oxidation'
NADHo.lower_bound=0
NADHo.upper_bound=1000

##### ========> AMP
AMPin=Reaction('AMPin')
AMPin.name='AMP import'
AMPin.lower_bound=1
AMPin.upper_bound=1

##### H <========> 
He=Reaction('He')
He.name='Proton exchange'
He.lower_bound=-1000
He.upper_bound=1000

##### H2O <========> 
H2Oe=Reaction('H2Oe')
H2Oe.name='Water exchange'
H2Oe.lower_bound=-1000
H2Oe.upper_bound=1000
```

```python
Gluc=Metabolite(
    'Gluc',
    formula='C6H12O6',
    name='Glucose',
    compartment='c')
ATP=Metabolite(
    'ATP',
    formula='C10H12N5O13P3',
    name='Adenosise tri-phosphate',
    compartment='c')
ADP=Metabolite(
    'ADP',
    formula='C10H15N5O10P2',
    name='Adenosise di-phosphate',
    compartment='c')
G6P=Metabolite(
    'G6P',
    formula='C6H13O9P',
    name='Glucose 6-phosphate',
    compartment='c')
H=Metabolite(
    'H',
    formula='H',
    name='Proton',
    compartment='c')
FDP=Metabolite(
    'FDP',
    formula='',
    name='Fructose 1,6-diphosphate',
    compartment='c')
F6P=Metabolite(
    'F6P',
    formula='',
    name='Fructose 6-phosphate',
    compartment='c')
DHAP=Metabolite(
    'DHAP',
    formula='',
    name='Dihydroxyacetone phosphate',
    compartment='c')
G3P=Metabolite(
    'G3P',
    formula='',
    name='Glyceraldehyde 3-phosphate',
    compartment='c')
NAD=Metabolite(
    'NAD',
    formula='',
    name='Nicotinamide adenine dinucleotide (oxidized)',
    compartment='c')
NADH=Metabolite(
    'NADH',
    formula='',
    name='Nicotinamide adenine dinucleotide (reduced)',
    compartment='c')
P=Metabolite(
    'P',
    formula='',
    name='Inorganic phosphate',
    compartment='c')
DPG13=Metabolite(
    'DPG13',
    formula='',
    name='1,3-Diphosphoglycerate',
    compartment='c')
PG3=Metabolite(
    'PG3',
    formula='',
    name='3-Phosphoglycerate',
    compartment='c')
PG2=Metabolite(
    'PG2',
    formula='',
    name='2-Phosphoglycerate',
    compartment='c')
PEP=Metabolite(
    'PEP',
    formula='',
    name='Phosphoenolpyruvate',
    compartment='c')
H2O=Metabolite(
    'H2O',
    formula='',
    name='Water',
    compartment='c')
PYR=Metabolite(
    'PYR',
    formula='',
    name='Pyruvate',
    compartment='c')
LAC=Metabolite(
    'LAC',
    formula='',
    name='Lactate',
    compartment='c')
AMP=Metabolite(
    'AMP',
    formula='',
    name='AMP',
    compartment='c')
```

```python
GLUin.add_metabolites({Gluc:1})
HK.add_metabolites({Gluc:-1,ATP:-1,G6P:1,ADP:1,H:1})
PGI.add_metabolites({G6P:-1,F6P:1})
PFK.add_metabolites({F6P:-1,ATP:-1,FDP:1,ADP:1,H:1})
TPI.add_metabolites({DHAP:-1,G3P:1})
ALD.add_metabolites({FDP:-1,DHAP:1,G3P:1})
GAPDH.add_metabolites({G3P:-1,NAD:-1,P:-1,DPG13:1,NADH:1,H:1})
PGK.add_metabolites({DPG13:-1,ADP:-1,PG3:1,ATP:1})
PGLM.add_metabolites({PG3:-1,PG2:1})
ENO.add_metabolites({PG2:-1,PEP:1,H2O:1})
PK.add_metabolites({PEP:-1,ADP:-1,H:-1,PYR:1,ATP:1})
LDH.add_metabolites({PYR:-1,NADH:-1,H:-1,LAC:1,NAD:1})
AMPe.add_metabolites({AMP:-1})
APK.add_metabolites({ADP:-2,AMP:1,ATP:1})
PYRe.add_metabolites({PYR:-1})
LACe.add_metabolites({LAC:-1})
ATPh.add_metabolites({ATP:-1,H2O:-1,ADP:1,P:1,H:1})
NADHo.add_metabolites({NADH:-1,NAD:1,H:1})
AMPin.add_metabolites({AMP:1})
He.add_metabolites({H:-1})
H2Oe.add_metabolites({H2O:-1})
```

```python
model.add_reactions([GLUin,HK,PGI,PFK,TPI,ALD,GAPDH,PGK,PGLM,ENO,PK,LDH,AMPe,APK,PYRe,LACe,ATPh,NADHo,AMPin,He,H2Oe])
```

```python
model.objective = 'PK'
```

```python
model.optimize()
```
